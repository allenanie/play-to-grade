```python
# Single Direction Agent & Random Agent
# & Accuracy      &  Precision   & Recall    & Accuracy  & Precision   & Recall

names = []

accuracy = []
accuracy_errs = []

precisions = []
precision_errs = []

recalls = []
recall_errs = []

rand_accuracy = []
rand_accuracy_errs = []

rand_precisions = []
rand_precision_errs = []

rand_recalls = []
rand_recall_errs = []

for row in text.split('\n'):
    cells = [c.strip() for c in row.split("&")]
    names.append(cells[0])
    cells = cells[1:]
    cells = [c.strip("\\").replace("$", "").split("\\pm") for c in cells]
    
    accuracy.append(float(cells[0][0]))
    accuracy_errs.append(float(cells[0][1]))
    
    precisions.append(float(cells[1][0]))
    precision_errs.append(float(cells[1][1]))
    
    recalls.append(float(cells[2][0]))
    recall_errs.append(float(cells[2][1]))
    
    rand_accuracy.append(float(cells[3][0]))
    rand_accuracy_errs.append(float(cells[3][1]))
    
    rand_precisions.append(float(cells[4][0]))
    rand_precision_errs.append(float(cells[4][1]))
    
    rand_recalls.append(float(cells[5][0]))
    rand_recall_errs.append(float(cells[5][1]))
```

```python
BIGGER_SIZE = 14
plt.figure(figsize=(8, 6))
plt.rc('font', size=BIGGER_SIZE)  # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)  # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)  # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)  # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)  # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)  # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.bar(np.array(range(len(names))) * 0.1 + 1 - 0.2, accuracy, width=0.08)
plt.bar(np.array(range(len(names))) * 0.1 + 2 - 0.2, precisions, width=0.08)
plt.bar(np.array(range(len(names))) * 0.1 + 3 - 0.2, recalls, width=0.08)

plt.xticks([1,2,3], ['Accuracy', 'Precision', 'Recall'])
plt.title("Single Direction Agent")
```
